```python

# classify ages into different groups
def age_groups(x):
    if 18 < x < 31:
        return "Young Adults"
    elif 30< x < 41:
        return "Junior Adults"
    elif 40< x < 51:
        return "Mid-age Adults"
    elif 50< x < 61:
        return "Senior Adults"
    elif 60< x < 71:
        return "The Old"
    else:
        return "The Elder"
df['age_group'] = df['age'].apply(age_groups)
```

```python
# Plot income distribution by age groups
plt.figure(figsize=(12,8))
order=['Young Adults', 'Junior Adults', 'Mid-age Adults', 'Senior Adults', 'The Old', 'The Elder']
sns.countplot(x = df['age_group'], hue = df['income'], order = order)
plt.title('Income distribution of different age groups', fontsize=20)
plt.legend()
```

1. In this figure, we can clearly see that the largest income gap is in `Young Adults`
2. People who are between 31 and 60 got the most >50K income amount, while the gap in `Junior Adults` still very big.

## All data have been cleaned.
## Next build model, compile model and predict model by tensor flow

```python
# Build a Logistic Regression model
num_trees = 100
max_features = 3
models = []
models.append(('LR', LogisticRegression()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('RF', RandomForestClassifier(n_estimators=num_trees, max_features=max_features)))
models.append(('XGB', XGBClassifier()))

# evalutate each model in turn
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=10, random_state=7)
    cv_results = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

Among those results, XGB got the best accuracy. I'll pick the XGB as the model

```python
xgb_model = XGBClassifier()
xgb_model.fit(x_train, y_train)
y_pred = xgb_model.predict(x_test)
print("Accuracy: %s%%" % (100*accuracy_score(y_test, y_pred)))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
